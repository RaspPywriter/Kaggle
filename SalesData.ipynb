```python
sales.columns = ['Invoice_ID', 'Branch', 'City', 'Customer_Types', 'Gender', 'Product_Line', 'Unit_Price', 'Quantity', 'Tax', 'Total', 'Date', 'Time', 'Payment', 'cogs', 'Gross_Margin_Percentage', 'Gross_Income','Rating']
```

**From describe(), the unit price (price of each product) varies from 10 to 99.99, with a mean of 55.67 and the standard deviation is 26.49. The quantity represents the number of products bought by each customer, with a minimum of 1 product, and a maximum of 10 products. For rating, which looks at customer satisfaction, the mean is 6.9.
I can see that Tax and Gross_Income are identical, therefore, I'll remove Gross_Income**

```python
sales = sales.drop(['Gross_Income'], axis=1)
```

**I will start by looking at the differnce between how many are members vs nonmembers and I will get the count of each type.**

```python
members = sales.groupby("Customer_Types")["Total"].count()
print(members)
```

**So from above, there are 501 members and 499 non-members.**

**Now get the sales by each group**

```python
totalSales = sales.groupby("Customer_Types")["Total"].sum()
print(totalSales)
```

**So the sales for members is 164,223 and sales for non-members is 158,743.**

**Get the average sales per customer type
Members spend, on average, 327.79 and non members spend 318.12**

```python
salesPercent = (totalSales/members)
print(salesPercent)
```

**Now let's see what members are buying**

```python
memberSales = sales[sales.Customer_Types == 'Member']
salesType = memberSales.groupby('Product_Line')['Customer_Types'].count()
print(salesType)
```

**The results from above show the count of categories that members bought. Food and beverage, Sports and travel, and fashion accessories represent the top three categories.**

**Next, I will look at the categories that non-members bought**

```python
nonMemberSales = sales[sales.Customer_Types == 'Normal']
salesTypeNonMember = nonMemberSales.groupby('Product_Line')['Customer_Types'].count()
print(salesTypeNonMember)
```

**The top three categories bought by non-members are: Electronic accessories, fashion accessories, and food and beverages. 
**Now I'll look at which branches the members shop at**

```python
salesMemberBranch = memberSales.groupby('Branch')['Customer_Types'].count()
print(salesMemberBranch)
```

**Members shop about evenly at all three branches**
**Next I will examine non-members, at each branch. Again it is very evenly split**

```python
salesNonMemberBranch = nonMemberSales.groupby('Branch')['Customer_Types'].count()
print(salesNonMemberBranch)
```

**Now I'll see the difference in what males vs females spend at the stores
Then I'll calculate the total of sales for each**

Of the top 10 sales, 4 customers gave ratings below 5, one was 5.2, one was 6.6, and 4 were above 7.2. **
